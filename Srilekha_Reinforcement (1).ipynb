Section A – Data Understanding & Cleaning (Pandas)  


```python
#1. Load the laptop.csv dataset into a Pandas DataFrame and display the first 5 rows.
import pandas as pd
import numpy as np
df=pd.read_excel("/content/laptopData.xlsx")
df.head(5)
```

```python
#2. Check the shape and data types of all columns.
df.shape
```

```python
#2. Check the shape and data types of all columns.
df.dtypes
```

```python
#3. Identify and handle missing values in the dataset. (Hint: use .isnull().sum() and  appropriate imputation).
(df.isnull().sum())
```

```python
#4. Standardize the values in the column OpSys (e.g., unify "Windows 10" and "Win  10").
df['OpSys'] = df['OpSys'].replace(['Windows 10 S', 'Windows 7'], 'Windows 10')
(df['OpSys'].unique())
```

```python
#5. Extract the numeric value from the column Ram (e.g., convert "8GB" → 8).
df['Ram']=df['Ram'].str.replace('GB','').astype(int)
df.head()
```

```python
#6. Split the column Memory into HDD and SSD capacity (Hint: use str.split).
df['HDD']=df['Memory'].str.split("HDD")
df['SSD']=df['Memory'].str.split("SSD")
df
```
